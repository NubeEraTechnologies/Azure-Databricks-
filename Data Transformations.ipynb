```python
data = [
    ("Alice", 34, "2024-01-05"),
    ("Bob", 45, "2024-01-07"),
    ("Cathy", None, "2024-01-10")
]

columns = ["name", "age", "join_date"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
df_filtered = df.filter(df.age > 35)
display(df_filtered)
```

```python
df_select = df.select("name", "age")
display(df_select)
```

```python
df_clean = df.na.drop(subset=["age"])
display(df_clean)
```

```python
from pyspark.sql.functions import to_date

df_cast = df.withColumn("join_date", to_date(df.join_date))
display(df_cast)
```

```python
from pyspark.sql.functions import col

df_new = df.withColumn("age_plus_5", col("age") + 5)
display(df_new)
```

```python
df_group = df.groupBy().avg("age")
display(df_group)
```

```python
data2 = [("Alice", "IT"), ("Bob", "HR")]
columns2 = ["name","department"]

df2 = spark.createDataFrame(data2, columns2)

df_join = df.join(df2, on="name", how="left")
display(df_join)
```
