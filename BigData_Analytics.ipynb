```python
data = [("North", 100), ("South", 200), ("West", 150)]
columns = ["region", "sales"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.filter(df.sales > 120).show()
```

```python
df.groupBy("region").sum("sales").show()
```

```python
from pyspark.sql.functions import col
df2 = df.withColumn("sales_x2", col("sales") * 2)
df2.show()
```

```python
df.write.format("delta").mode("overwrite").save("/tmp/bigdata/sales_delta")
```

```python
spark.read.format("delta").load("/tmp/bigdata/sales_delta").show()
```
